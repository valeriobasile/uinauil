# Metadata

This notebook shows the values of the **metadata** for each task available on the `uinauil` module.

```python
from evalita4elg.uinauil.src import uinauil as ul
# import uinauil as ul
```

```python
for task_name in ul.tasks:
    task = ul.Task(task_name)
    print(f"### task name: {task.task_name}")
    print(f"- desc sample:\n{task.desc[:300]}...")
    print(f"- link: {task.link}")
    print(f"- data.keys: {task.data.keys}")
    print(f"- data.feature_keys: {task.data.feature_keys}")
    print(f"- data.feature_dim: {task.data.feature_dim}")
    print(f"- data.feature_info: {task.data.feature_info}")
    if len(task.data.feature_keys) != len(task.data.feature_dim):
        print(f"[ERROR] data.feature_key and data.feature_dim must have the same size")
    print(f"- data.target_key: {task.data.target_key}")
    print(f"- data.target_dim: {task.data.target_dim}")
    print(f"- data.target_values: {task.data.target_values}")
    print(f"- data.target_desc: {task.data.target_desc}")
    if len(task.data.target_values) != len(task.data.target_desc):
        print(f"[ERROR] data.target_values and data.target_desc must have the same size")
    train = task.data.training_set
    print(f"- data.training_set ({len(train)} entries) sample:\n{train[:2]}")
    test = task.data.test_set
    print(f"- data.test_set ({len(test)} entries) sample:\n{test[:2]}")
    print("\n############################################################################\n")
    
```
